```python
threshold2 = 2772
```

```python
df_threshtimesm1 = m1alldf100wt.groupby('simulation').apply(lambda x: (x > threshold2).idxmax()).apply(lambda x: [y[1] for y in x])
df_threshtimesm2   = m2alldf100wt.groupby('simulation').apply(lambda x: (x > threshold2).idxmax()).apply(lambda x: [y[1] for y in x])
df_threshtimesm3 = m3alldf100wt.groupby('simulation').apply(lambda x: (x > threshold2).idxmax()).apply(lambda x: [y[1] for y in x])
df_threshtimesm4   = m4alldf100wt.groupby('simulation').apply(lambda x: (x > threshold2).idxmax()).apply(lambda x: [y[1] for y in x])
```

```python
clus0n = pd.DataFrame({'WT1': df_threshtimesm1['MLKLa_obs'], 
                       'WT2': df_threshtimesm2['MLKLa_obs'],
                      'WT3': df_threshtimesm3['MLKLa_obs'], 
                       'WT4': df_threshtimesm4['MLKLa_obs']})
dfclus0n = pd.DataFrame(data = clus0n, columns = ['WT1', 'WT2','WT3', 'WT4'])
dfclus0n = pd.melt(dfclus0n)
```

```python
import seaborn as sns
sns.set(style="whitegrid")
plt.figure(figsize = (15,10))
my_pal = {"WT1": "indianred", "WT2": "lightsalmon", "WT3": "mediumseagreen", "WT4":"cornflowerblue"}
# my_pal1 = {"wt0": "red", "a200":"blue"}
# b = sns.boxplot(x = "value", y ="variable", data = dfclus0, notch = True, width = 0.5, palette = my_pal)
b = sns.swarmplot(x="value", y="variable", data=dfclus0n, zorder=2, palette=my_pal)
# b = sns.boxplot(x = "value", y ="variable", data = dfclus0, notch = True, width = 0.5, palette = my_pal)
# b = sns.swarmplot(x="value", y="variable", data=dfclus0, zorder=2, palette=my_pal1)
b.set_xlabel("Time (m)",fontsize=15)
# b.set_ylabel("Mode 1 Degeneracy",fontsize=15)
b.tick_params(labelsize=15)
plt.xlim(xmax = 1800, xmin = 0)
# plt.savefig('figs/Cluster0/pmlkl_100tnf_clus0box_wswarm_5720_modnew.pdf')
plt.show()
```
